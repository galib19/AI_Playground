```python
import torch
```

```python
x = torch.tensor(1.0)
y = torch.tensor(2.0)

w = torch.tensor(1.0, requires_grad=True)
```

```python
#forward pass and loss
y_hat = w*x
loss = (y_hat-y)**2

print(loss)
```

```python
#backward pass

loss.backward()
print(w.grad)
```
