```python
import torch
```

```python
x = torch.randn(3, requires_grad=True)
x
```

```python
y = x+2
y
```

```python
z = y*y*2
z
```

```python
z=z.mean()
z
```

```python
z.backward() #dz/dx
x, x.grad
```

```python
z2 = y*y*2
# z2.backward() #when z2 is not scalar, backward() will not work
#because in the background it is a vector jacobian product
#to make it working...have to pass vector argument to the backward()

v = torch.tensor([0.1, 1.0, 0.001], dtype = torch.float32)
z2.backward(v)
x, x.grad
```

```python
#how to stop gradient tracking off
# 3 ways:
# x.requires_grad_(False)
# x.detach()
# with torch.no_grad():

x.requires_grad_(False)
print(x)
```

```python
x = torch.randn(3, requires_grad=True)
y = x.detach()
x, y
```

```python
with torch.no_grad():
  y = x+2
  print(y)
print(y), print(x)
```

```python
weights = torch.ones(4, requires_grad=True)
for epoch in range(3):
  model_output = (weights*3).sum()
  model_output.backward()
  print(weights.grad)
  #gradients are wrong
```

```python
weights = torch.ones(4, requires_grad=True)
for epoch in range(3):
  model_output = (weights*3).sum()
  model_output.backward()
  print(weights.grad)
  weights.grad.zero_()
```

```python
weights = torch.ones(4, requires_grad=True)

optimizer = torch.optim.SGD(weights, lr = 0.01)
optimizer.step()
optimizer.zero_grad() # will do the same as previous cell
```
